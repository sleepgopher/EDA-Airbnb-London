```python
# Importing Libraries

import glob
#import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
# Reading Files and Creating Data Frame

listing = pd.read_csv("http://data.insideairbnb.com/united-kingdom/england/london/2020-05-10/visualisations/listings.csv")
#reviews = pd.read_csv("/Users/yalinyener/GitHub/yalinyener/EDA-Airbnb-London/data/reviews.csv")
#calendar = pd.read_csv("/Users/yalinyener/GitHub/yalinyener/EDA-Airbnb-London/data/calendar.csv")
#detail_listing = pd.read_csv("/Users/yalinyener/GitHub/yalinyener/EDA-Airbnb-London/data/listings-2.csv")
#detail_reviews = pd.read_csv("/Users/yalinyener/GitHub/yalinyener/EDA-Airbnb-London/data/reviews-2.csv")
house_price = pd.read_csv("/Users/yalinyener/GitHub/yalinyener/EDA-Airbnb-London/data/HousePrice.csv")
```

```python
#Checking Data Frame - listing

listing.sample(5)
```

```python
#Checking DataFrame size

listing.shape
```

```python
#Checking general info about dataframe

listing.info()
```

```python
# Checking columns names

listing.columns
```

```python
#Checking duplicates rows and drop.

listing.drop_duplicates(subset=None, keep="first", inplace=True)
```

```python
#Checking null values based on columns

listing.isnull().sum()
```

```python
# Deleting full na column

del listing["neighbourhood_group"]
```

```python
#Checking basic descriptive statistics for numerical columns.

listing.describe()
```

```python
#Brief Insight of listing

print("Number of properties:")
print(len(listing["id"]))

print("Number of unique host:")
print(len(listing["host_name"].unique()))

print("Number of Room Type:")
print(len(listing["room_type"].unique()))

print("Number of Borough:")
print(len(listing["neighbourhood"].unique()))

print("Average Price for All London:")
print(round(listing.price.mean(),2))

print("Number of 0 (Zero) Price:")
print(len(listing[listing["price"]==0]))

print("Average Number of Revies for All London:")
print(round(listing.number_of_reviews.mean(),2))
```

```python
#Deleting rows that price is equal zero

listing = listing[listing["price"]>0]
len(listing)
```

```python
#Group by Room Type 

room_type_group_count = listing.groupby("room_type").id.count().sort_values(ascending=False)
room_type_group_count = room_type_group_count.reset_index()
room_type_group_count.rename(columns={"id":"count"}, inplace=True)
room_type_group_count
```

```python
#Barplot of Room Type 

plt.figure(figsize=(15,10),dpi=200),
plt.style.use('default')
ax = sns.barplot(x="room_type", y="count", data=room_type_group_count)
plt.title("Disturibution of Room Type", weight="bold", c="red", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=0, weight="bold")
plt.ylabel("Number of Properties)",weight= "bold",c="k")
plt.yticks(weight="bold")
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                    ha='center', va='bottom',weight="bold",
                    color= 'black')
plt.savefig('images/01_DistuributionofRoomType.png', dpi=200)

plt.show()
```

```python
# Group by Borough for count

borough_group_count = listing.groupby("neighbourhood").id.count().sort_values(ascending=False)
borough_group_count = borough_group_count.reset_index()
borough_group_count.rename(columns={"id":"count"}, inplace=True)
borough_group_count.head()
```

```python
#Barplot of Borough 

plt.figure(figsize=(15,10),dpi=200),
plt.style.use('default')
ax = sns.barplot(x="neighbourhood", y="count", data=borough_group_count)
plt.title("Disturibution based on Borough", weight="bold", c="red", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=90, weight="bold")
plt.ylabel("Number of Properties)",weight= "bold",c="k")
plt.yticks(weight="bold")
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                    ha='center', va='bottom',
                    color= 'black', size=8)
plt.savefig('images/02_DistuributionofBorough.png', dpi=200)

plt.show()
```

```python
# Group by Borough for price

borough_group_price = listing.groupby("neighbourhood").price.mean().sort_values(ascending=False)
borough_group_price = borough_group_price.reset_index()
borough_group_price.tail()
```

```python
# Group by Borough for average number of reviews

borough_group_reviews = listing.groupby("neighbourhood").number_of_reviews.mean().sort_values(ascending=False)
borough_group_reviews = borough_group_reviews.reset_index()
borough_group_reviews.head()
```

```python
# Number of properties based on Borough that over 100 reviews

over_100_reviews = listing.sort_values("number_of_reviews",ascending=False)
over_100_reviews = over_100_reviews[over_100_reviews["number_of_reviews"]>=100]
over_100_reviews = over_100_reviews.groupby("neighbourhood").count().sort_values("number_of_reviews",ascending=False).head(10)
over_100_reviews
```

```python
# Barplot of properties based on Borough that over 100 reviews

plt.figure(figsize=(15,10),dpi=200),
plt.style.use('default')
ax = sns.barplot(x=over_100_reviews.index, y="number_of_reviews", data=over_100_reviews)
plt.title("Number of Properties that over 100 Reviews", weight="bold", c="red", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=90, weight="bold")
plt.ylabel("Number of Properties",weight= "bold",c="k")
plt.yticks(weight="bold")
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                    ha='center', va='bottom',
                    color= 'black', size=8, weight="bold")
plt.savefig('images/04_NumberOfPropertiesOver100PerBorough.png', dpi=200)

plt.show()
```

```python
# Property Type and Borough

property_type_borough = listing.groupby(["neighbourhood","room_type"]).count().reset_index()
property_type_borough = property_type_borough.pivot("room_type","neighbourhood","id")
property_type_borough
```

```python
#Heatmap of Property Type and Borough

plt.figure(figsize=(15,16),dpi=200),
sns.set(font_scale=1)
sns.heatmap(property_type_borough,annot=False,fmt="d", linewidths=2, cmap="Oranges",cbar_kws={'label': 'Number of Properties','orientation':'horizontal'})
plt.style.use('default')
plt.ylabel("")
plt.xlabel("")
plt.yticks(weight="bold",rotation=0, size=12)
plt.xticks(weight="bold",rotation=90,size=8)
plt.title("Heatmap of Room Type and Borough",weight="bold",c="red", fontsize=15)
plt.savefig('images/05_PropertyTypeBoroughCountHeatmap.png', dpi=200)
plt.show()
```

```python
listing.head()
```

```python
#Availability Per Borough

availability_borough = listing.groupby("neighbourhood").availability_365.mean().sort_values().reset_index()
availability_borough.head()
```

```python
# Barplot of #Availability Per Borough
availability_borough = availability_borough.head()
plt.figure(figsize=(15,10),dpi=200),
plt.style.use('default')
ax = sns.barplot(x="neighbourhood", y="availability_365", data=availability_borough)
plt.title("Availability of 365 Per Borough", weight="bold", c="red", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=90, weight="bold")
plt.ylabel("Average Price",weight= "bold",c="k")
plt.yticks(weight="bold")
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                    ha='center', va='bottom',
                    color= 'black', size=12, weight="bold")
plt.savefig('images/05_Availabilityof365PerBorough.png', dpi=200)

plt.show()
```

```python
#Calculating income_yearly

borough_ind=listing.groupby("neighbourhood").mean().reset_index()
borough_ind = borough_ind[["neighbourhood","price","availability_365"]]
borough_ind["occupancy"] = 365 - borough_ind["availability_365"]
borough_ind["income_yearly"] = borough_ind["occupancy"] * borough_ind["price"]
borough_ind.sort_values("income_yearly", ascending=False).head(10)
```

```python
# Barplot of Availability Per Borough

borough_ind_graph = borough_ind.sort_values("income_yearly", ascending=False).head(10)
plt.figure(figsize=(15,10),dpi=200),
plt.style.use('default')
ax = sns.barplot(x="neighbourhood", y="income_yearly", data=borough_ind_graph)
plt.title("Annual Income Per Borough", weight="bold", c="red", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=90, weight="bold")
plt.ylabel("Annual Income (£) ",weight= "bold",c="k")
plt.yticks(weight="bold")
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                    ha='center', va='bottom',
                    color= 'black', size=12, weight="bold")
plt.savefig('images/06_AnnualIncomePerBorough.png', dpi=200)

plt.show()
```

```python
house_price = house_price[["Borough","Average"]]
house_price.sort_values("Average").head()
```

```python
borough_ind.head()
```

```python
#Calculating turnaround value
borough_price_ind = borough_ind.merge(house_price, left_on='neighbourhood', right_on='Borough')
borough_price_ind = borough_price_ind[["neighbourhood","price","occupancy","income_yearly","Average"]]
borough_price_ind["Average"] = borough_price_ind["Average"].str.replace(",","")
borough_price_ind["Average"] = borough_price_ind["Average"].astype("int")
borough_price_ind["turnaround"] = borough_price_ind["Average"] / borough_price_ind["income_yearly"]
borough_price_ind.columns = ['borough', 'price/night', 'occupancy_day', 'annual_income', 'average',
       'turnaround']
borough_price_ind_sorted = borough_price_ind.sort_values("turnaround").set_index("borough")
borough_price_ind_sorted.head()
```

```python
# Barplot of Turnaround Rate
borough_price_ind_sorted_graph = borough_price_ind_sorted.head(10)
plt.figure(figsize=(15,10),dpi=200),
plt.style.use('default')
ax = sns.barplot(borough_price_ind_sorted_graph.index, y="turnaround", data=borough_price_ind_sorted_graph)
plt.title("Turnaround Rate of Borough", weight="bold", c="red", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=90, weight="bold")
plt.ylabel("Turnaround Rate (year) ",weight= "bold",c="k")
plt.yticks(weight="bold")
plt.savefig('images/07_TurnaroundRatePerBorough.png', dpi=200)

plt.show()
```

```python
# Barplot of Average House Price Rate
plt.figure(figsize=(15,10),dpi=200),
plt.style.use('default')
ax = sns.barplot(x="borough", y="average", data=borough_price_ind.sort_values("average"))
plt.title("House Price of Borough", weight="bold", c="red", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=90, weight="bold")
plt.ylabel("House Price (£) ",weight= "bold",c="k")
plt.yticks(weight="bold")
plt.savefig('images/08_HousePricePerBorough.png', dpi=200)
plt.show()
```

```python
#Scale all columns values beetween 0 and 1

scaler = MinMaxScaler()

borough_price_ind_sorted[['price/night', 'occupancy_day','annual_income','average','turnaround']] = scaler.fit_transform(borough_price_ind_sorted[['price/night', 'occupancy_day','annual_income','average','turnaround']])
borough_price_ind_sorted["turnaround"] = 1 - borough_price_ind_sorted["turnaround"]

borough_price_ind_sorted
```

```python
#Heatmap of Property Type and Borough
plt.figure(figsize=(20,16),dpi=200),
sns.set(font_scale=1)
sns.heatmap(borough_price_ind_sorted,annot=True, fmt=".2f", linewidths=2, cmap="Oranges",cbar_kws={'label': 'Scale','orientation':'horizontal'})
plt.style.use('default')
plt.ylabel("")
plt.xlabel("")
plt.yticks(weight="bold",rotation=0, size=12)
plt.xticks([0.5,1.5,2.5,3.5,4.5], ['Price/Night', 'Occupancy', 'Income','House Price','Turnaround'],weight="bold",rotation=0,size=12)
plt.title("Heatmap of Borough's Features",weight="bold",c="red", fontsize=15)
plt.savefig('images/09_BoroughFeatureCountHeatmap.png', dpi=200)
plt.show()
```

```python
#Calculating productivity point for each borough

borough_price_ind_sorted["averageT"] = 1- borough_price_ind_sorted["average"]


borough_price_ind_sorted["total"] = borough_price_ind_sorted["price/night"]+borough_price_ind_sorted["occupancy_day"]+borough_price_ind_sorted["annual_income"]+borough_price_ind_sorted["averageT"]+borough_price_ind_sorted["turnaround"]
borough_price_ind_sorted.sort_values("total",ascending=False).head(10)
```

```python
#Barplor for productivity point
borough_price_ind_sorted_graph = borough_price_ind_sorted.sort_values("total",ascending=False)
plt.figure(figsize=(15,10),dpi=200),
plt.style.use('default')
ax = sns.barplot(borough_price_ind_sorted.index, y="total", data=borough_price_ind_sorted_graph)
plt.title("London Borough's Productivity for Investment", weight="bold", c="red", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=90, weight="bold")
plt.ylabel("Productivity",weight= "bold",c="k")
plt.yticks(weight="bold")
for p in ax.patches:
        ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                    ha='center', va='bottom',
                    color= 'black', size=8, weight="bold")
plt.savefig('images/10_LondonBoroughsFeatureBarPlot.png', dpi=200)

plt.show()
```
